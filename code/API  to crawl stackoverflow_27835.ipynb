```
from datetime import datetime
import calendar

def dateToTimestamp(x):
  date = datetime.strptime(x,"%Y/%m/%d")
  timestamp = calendar.timegm(date.utctimetuple())
  return timestamp
```

```
question_id = []
for i in range(len(questions['items'])):
  # print(questions['items'])
  if(questions['items'][i].get('is_answered') == True and questions['items'][i].get('accepted_answer_id') != None):
    question_id.append(questions['items'][i].get('question_id'))
```

```
import csv
import numpy as np
result = np.array(question_id)
with open('0101_0115.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    for row in range(0, result.shape[0]):
        myList = []
        myList.append(result[row])
        writer.writerow(myList)
```
