```python
import re
question_readability = [] 
question_body = []
for i in range(len(df['question_body'])):
  rex = re.compile(r'<code.*?>(.*?)</code>',re.S|re.M)
  s = df['question_body'][i]
  s = s.replace('\n',"")
  s = s.replace('\r',"")
  s = s.replace('\t',"")
  match = rex.findall(s)
  word_cnt = 0
  for i in range(len(match)):
    word_cnt += len(match[i].split())
  question_readability.append(word_cnt)
  tmp = re.sub(r'<code.*?>(.*?)</code>', '', s)
  # print(tmp)
  # print(type(tmp))
  clean = re.compile('<.*?>')
  question_body.append(re.sub(clean, '',tmp))
```

```python
df['question_code_size'] = question_readability
df['question_body'] = question_body
```
